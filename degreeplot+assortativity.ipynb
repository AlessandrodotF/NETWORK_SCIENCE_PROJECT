## Degree distribution

```python
def plot_degree_distr(G):
    params = {'mathtext.default': 'regular',
              "axes.titlesize" : "24",
              "axes.titlesize" : "18",
              "axes.labelsize" : "20",
              "xtick.labelsize" : "16",
              "ytick.labelsize" : "16"}          
    plt.rcParams.update(params)
    degrees = [val for (node, val) in G.degree()]
    sort_degrees = sorted(degrees)
    degrees_hist = [degrees.count(x) for x in sort_degrees]

    fig, (ax1, ax2) = plt.subplots(1, 2,figsize=(15,10))
    fig.suptitle('Degree distribution',fontsize=15)
    
    ax1.plot(sort_degrees, degrees_hist, '.', linestyle="None")
    ax1.grid(True)
    ax1.set_title("Linear scale")
    ax1.set_xlabel("k")
    ax1.set_ylabel("$p_{k}$")
    
    
    plt.xlim([0, 1*10**5])
    
    ax2.loglog(sort_degrees, degrees_hist, '.', linestyle="None")
    ax2.grid(True)
    ax2.set_title("LogLog scale")
    ax2.set_xlabel("k")
    ax1.set_ylabel("$p_{k}$")
    
    plt.xlim([1, 1*10**5])
    plt.show()
    plt.close()
        
    return
```

```python
plot_degree_distr(G)
```

## Gamma estinamation

```python
def compute_gamma(G):
    degrees = [val for (node, val) in G.degree()]
    fit = powerlaw.Fit(degrees)
    gamma = fit.alpha
    return gamma
```

```python
gamma=compute_gamma(G)
print(gamma)
```

## Assortativity: copy and paste from wiki

The assortativity coefficient is the Pearson correlation coefficient of degree between pairs of linked nodes.

The assortativity coefficient is given by 
${\displaystyle r={\frac {\sum _{jk}{jk(e_{jk}-q_{j}q_{k})}}{\sigma _{q}^{2}}}}$

Positive values of r indicate a correlation between nodes of similar degree, while negative values indicate relationships between nodes of different degree. In general, r lies between −1 and 1. When r = 1, the network is said to have perfect assortative mixing patterns, when r = 0 the network is non-assortative, while at r = −1 the network is completely disassortative.
